Import the Libraries

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
# Loading train and test data
trainData =  pd.read_csv('train_LR.csv')
trainData.head()
```

```python
# Creating the dependent and independent variables for the training module(X_train, y_train)
y_train = trainData['CycleBuyer']
X_train = trainData.drop('CycleBuyer', axis=1)
X_train.head()
```

```python
y_train.head()
```

```python
testData = pd.read_csv('test_LR.csv')
testData.head()
```

```python
y_test = testData['CycleBuyer']
X_test = testData.drop('CycleBuyer', axis=1)
X_test.head()
```

```python
y_test.head()
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
y_pred_train = model.predict(X_train)
y_pred_train
```

```python
y_pred_test = model.predict(X_test)
y_pred_test
```

```python
# Creating classification report
classificationReport = classification_report(y_test,y_pred_test)
print("Classification report \n",classificationReport)
```

```python
# Creating confusion matrix
confusion_matrix(y_test,y_pred_test)
```

```python
tn, fp, fn, tp = confusion_matrix(y_test, y_pred_test).ravel()
(tn,fp, fn, tp)
```
